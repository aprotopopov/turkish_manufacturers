### Отчет по экономической активности Турции
-------

В данном отчете представлено исследование по экономической географии Турции с помощью выделения географических кластеров с преобладающими формами экономической активности. 

Для сбора данных использовался сайт http://www.turkish-manufacturers.com, в котором собраны данные о компаниях работающих в Турции с разделением на категории, подкатегории и продукты. К каждому продукту относится подможество фирм, для различных продуктов эти подмножества могут пересекаться. Для скачивания данных использовались пакеты `python` `requests`, `BeatifulSoup`. Скрипты работали паралелльно с помощью библиотеки `multiprocessing`. Необходимые функции находятся в файле `get_turkish_manufacturers.py`. Сначала предполагался обход все ссылок и скачивание данных по компаниям для каждого продукта, но вследствие большого количество дубликатов было решено выделять только уникальные названия компаний по ссылкам. При этом предполагается, что одна фирма не может быть в разных категориях, предположение оказалось неверным, и пришлось использовать уникальные фирмы в пределах одной категории. Продукты в разных категориях отличаются, в пределах одной могут повторяться, что исключается контролем уникальности продуктов `seen_products`. Далее приводится код, для получения данных. **Выполнение занимает значительное время (может быть до нескольких суток)**. Готовые данные можно извлечь, пропуская второй и третий блоки кода (с закачкой и объединением).

```python
%matplotlib inline 
import requests
import os
import gc
import itertools as it
from bs4 import BeautifulSoup
from multiprocessing.pool import ThreadPool as Pool

import warnings
# to suppress PerformanceWarning from df.to_hdf fixed
warnings.filterwarnings('ignore')
```

```python
from get_turkish_manufacturers import (get_sublinks, get_categories, 
    get_products, get_firms, get_firm_pages, get_firms_uniq)

# if not first run to skip already processed products   
# dfs = []
# for file in os.listdir():
#     if file.endswith('.csv'):
#         dfs.append(pd.read_csv(file))
# seen_product_np = big_df['product'].unique()
# np.savetxt('seen_products_file', seen_product_np, fmt='%s')

# if os.path.exists('seen_products_file'):
#     seen_file = pd.read_table('seen_products_file', header=None)
#     seen_products = set(seen_file.values.flatten())


seen_products = set()


if not os.path.exists('results'):
    os.mkdir('results')

url = requests.get('http://www.turkish-manufacturers.com/')
url_pref = 'http://www.turkish-manufacturers.com'
soup = BeautifulSoup(url.text, 'lxml')
links_main = soup.find_all('a', class_='slink')
url_categories, name_categories = get_sublinks(links_main)

iteration = 0
num_workers = 20
pool = Pool(num_workers)

for url_category, name_category in zip(url_categories, name_categories):
    url_subcategories, name_subcategories = get_categories(url_category)
    
    seen_firms = set()
    for url_subcategory, name_subcategory in zip(url_subcategories, name_subcategories):
        url_products, name_products, seen_products = get_products(url_subcategory, seen_products)

        iteration += 1
        if not url_products:
            continue

        product_firms = []
        for url_product, name_product in zip(url_products, name_products):
            firm_pages = get_firm_pages(url_product)

            firms_final = pool.starmap(get_firms_uniq, list(zip(firm_pages, it.repeat(seen_firms))))
            firms_list = list(it.chain(*firms_final))
            firms_with_prd_name = [l + [name_product] for l in firms_list]
            product_firms.extend(firms_with_prd_name)

        print('iteration {} is finished, numbers of firms {}'.format(iteration, len(product_firms)))
        print('{}/{}'.format(name_category, name_subcategory))

        # df = pd.DataFrame(firms_list, columns=['name', 'firm_link', 'address', 'keys', 'product', 'url_located'])
        df = pd.DataFrame(firms_list, columns=['name', 'firm_link', 'address', 'keys', 'product', 'url_located'])
        df['category'] = name_subcategory
        df['field'] = name_category

        # csv was suppressed due to utf-8 encoding error
        df_name = '{}__{}.h5'.format(name_category, name_subcategory)
        # df_name = '{}__{}.csv'.format(name_category, name_subcategory)
        df_name = df_name.replace(' ', '_')
        df_file_path = 'results/{}'.format(df_name)
        df_cols = ['name', 'firm_link', 'address', 'keys', 'product', 
        'url_located','category', 'field']

        if os.path.exists(df_file_path):
            # df_existed = pd.read_csv(df_file_path)
            df_existed = pd.read_hdf(df_file_path)
            df = pd.concat([df_existed, df])

        # df.to_csv(df_file_path)
        df[df_cols].to_hdf(df_file_path, 'df', complib='bzip2', complevel=9, mode='w')

        del df
        gc.collect()
```

Основная идея скрипта скачать данные о названии, адресе фирм, ключевых словах, остальные поля вспомогательные. Результаты сохраняются в `pandas DataFrame` в формате `.hdf` для каждой категории и подкатегории, при условии, что в подкатегории есть продукты, ранее не встреченные при обходе. ## Кластеризация

После сбора данных, выделения топонимов и получения широты и долготы необходимо провести географическую кластеризацию. Для этих целей подходит алгоритм [DBSCAN](https://en.wikipedia.org/wiki/DBSCAN), который может применяться к данным с шумами, также он способен выявлять заранее неизвестное число кластеров. Этот алгоритм реализован в библиотеке [scikit-learn](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.dbscan.html). При поиске необходимого алгоритма также были рассмотрены [KMeans](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html) и [HDBSCAN](https://github.com/lmcinnes/hdbscan). Для отображения кластеров на карте использовался [Matplotlib Basemap Toolkit](http://matplotlib.org/basemap/index.html). Т.к. не удалось собрать все данные за отведеный срок, то полноценной картины представить не удалось. Подготовим данные:

```python
import ast
X = np.array(df.lat_lng.tolist())

# to skip false lattitude and longitude
mask_lat = (X[:, 0] < 43) & (X[:, 0] > 35)
mask_lng = (X[:, 1] < 45) & (X[:, 1] > 25)

X = X[mask_lat & mask_lng]
# swap lat and lng for visialization in maps
X[:,[0, 1]] = X[:,[1, 0]]
```

Для начала попробуем параметры по умолчанию для DBSCAN, объявим функцию для кластеризации. Функция с подгрузкой карты с помощью Basemap, если этот модуль отсутствует, то он будет пропущен. Также объявлена функция с прорисовкой кластеризации DBSCAN.

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

def fit_dbscan(X, eps=0.1, min_samples=10):
    # Compute DBSCAN
    db = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    print('Estimated number of clusters: %d' % n_clusters_)
    return labels, core_samples_mask

def add_map():
    try:
        from mpl_toolkits.basemap import Basemap
        fig = plt.figure(figsize=(15,10))
        ax = fig.add_axes([0,0,1,1])
        m = Basemap(projection='cyl', llcrnrlon=25.,llcrnrlat=35.,urcrnrlon=45, urcrnrlat=43., resolution='i', ax=ax);
        m.drawcoastlines(linewidth=1);
        m.drawcountries(linewidth=1);
        m.drawmapboundary(fill_color='lightblue');
        m.fillcontinents(color='#ddaa66',lake_color='lightblue');
    except ImportError:
        fig = plt.figure(figsize=(15,10))

def plot_clusters(X, labels, core_samples_mask):
    # Black removed and is used for noise instead.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    unique_labels = set(labels)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = 'k'

        class_member_mask = (labels == k)

        xy = X[class_member_mask & core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=14)

        xy = X[class_member_mask & ~core_samples_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=6)

    plt.title('Estimated number of clusters: %d' % n_clusters_); 
```

Теперь применим DBSCAN и визуализируем результат.

```python
labels, core_samples_mask = fit_dbscan(X, eps=0.1, min_samples=10)
add_map()
plot_clusters(X, labels, core_samples_mask)
```

Теперь попробуем немного увеличить `eps`:

```python
labels, core_samples_mask = fit_dbscan(X, eps=0.2, min_samples=10)
add_map()
plot_clusters(X, labels, core_samples_mask)
```

И немного уменьшить относительно исходного:

```python
labels, core_samples_mask = fit_dbscan(X, eps=0.05, min_samples=10)
add_map()
plot_clusters(X, labels, core_samples_mask)
```

Основная проблема DBSCAN - это его зависимость от `eps`, и `min_samples`. Решением этой проблемы может служить алгоритм HDBSCAN, который зависит только от одного параметра `min_cluster_size`. Работает похоже на DBSCAN, но в процессе решения подбирает различные `eps` и ищет результат, наиболее стабильный относительно `eps`. Попробуем HDBSCAN, для начала объявим функцию для применения алгоритма:

```python
from hdbscan import HDBSCAN
def fit_hdbscan(X, min_cluster_size=10):
    # Compute DBSCAN
    hdb = HDBSCAN(min_cluster_size=min_cluster_size).fit(X)
    labels = hdb.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

    print('Estimated number of clusters: %d' % n_clusters_)
    return labels

def plot_hdbscan(X, labels):
    # Black removed and is used for noise instead.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    unique_labels = set(labels)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = 'k'
            markersize = 6
        else:
            markersize = 14
        
        class_member_mask = (labels == k)

        xy = X[class_member_mask]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
                 markeredgecolor='k', markersize=markersize)

       
    plt.title('Estimated number of clusters: %d' % n_clusters_); 
```

Посмотрим на результат применения этого алгоритма:

```python
labels = fit_hdbscan(X, min_cluster_size=10)
add_map()
plot_hdbscan(X, labels)
```

По результатам видно, что hdbscan находит около 500 кластеров при размере кластера 10. Поэтому необходимо увеличить его, допустим до 100:

```python
labels = fit_hdbscan(X, min_cluster_size=100)
add_map()
plot_hdbscan(X, labels)
```

При данных параметрах HDBSCAN находит более ярко выраженные кластеры, при это все равно довольно много шумов в районе Стамбула и Анкары. В пределах этих городов выделено несколько кластеров, данный масштаб не позволяет это отобразить.
